
# Sensor Temperature Analysis

This notebook analyzes daily temperature readings from 4 weather stations using NumPy.

## Tasks:
1. Print the shape and compute mean temperature per station.
2. Extract temperatures above 28°C using boolean masking.
3. Normalize the data to range [0, 1].


```

import numpy as np

data = np.array([[22.5, 19.0, 31.2, 28.7, 25.1],
                 [17.3, 22.8, 30.5, 26.4, 21.9],
                 [33.1, 29.6, 18.4, 24.0, 27.8],
                 [20.2, 23.5, 31.9, 28.1, 22.6]])

data
```

```

# Task 1: Shape and Mean per Station

print("Shape of data:", data.shape)

mean_per_station = data.mean(axis=1)
print("Mean temperature per station:", mean_per_station)
```

```

# Task 2: Extract temperatures above 28°C

high_temps = data[data > 28.0]
print("Temperatures above 28°C:", high_temps)
```

```

# Task 3: Normalize data to range [0, 1]

normalized = (data - data.min()) / (data.max() - data.min())
normalized_rounded = np.round(normalized, 2)

print("Normalized Data:\n", normalized_rounded)
```
