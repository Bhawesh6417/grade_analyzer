## Import Required Libraries

```

import pandas as pd
import numpy as np
```

## Create the Patient DataFrame

```

data = {
    'patient_id': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110,
                   111, 112, 113, 114, 115, 101, 107, 118, 119, 120],
    'age': ['25', '34', None, '45', '29', None, '38', '52', '27', '41',
            '33', 'unknown', '48', '26', '35', '25', '38', '31', None, '44'],
    'weight': ['70', '65', '80', None, '75', None, '68', '90', '72', '85',
               '78', None, '82', '69', 'N/A', '70', '68', '74', None, '88'],
    'blood_pressure': [120, 130, None, 140, 125, None, 135, None, 118, 145,
                      128, None, 138, 122, None, 120, 135, 126, None, 142],
    'medication': ['Aspirin', 'Metformin', 'Lisinopril', None, 'Aspirin',
                   'Metformin', 'Lisinopril', 'Aspirin', None, 'Metformin',
                   'Lisinopril', 'Aspirin', None, 'Metformin', 'Aspirin',
                   'Aspirin', 'Lisinopril', 'Metformin', 'Aspirin', None],
    'insurance_provider': ['Blue Cross', 'Aetna', 'Cigna', 'UnitedHealth', None,
                          'Blue Cross', 'Aetna', 'Cigna', 'UnitedHealth', 'Blue Cross',
                          'Aetna', None, 'UnitedHealth', 'Blue Cross', 'Aetna',
                          'Blue Cross', 'Aetna', 'Cigna', 'UnitedHealth', None]
}

df = pd.DataFrame(data)
df
```

## === TASK 1: Inspect the Data ===

```

print("Data Info:")
df.info()

print("\nMissing Values Count:")
print(df.isnull().sum())

print("\nMissing Percentage:")
print((df.isnull().sum() / len(df)) * 100)

print("\nDuplicate Rows:", df.duplicated().sum())
```

## === TASK 2: Data Type Conversion ===

```

# Convert age and weight to numeric
df['age'] = pd.to_numeric(df['age'], errors='coerce')
df['weight'] = pd.to_numeric(df['weight'], errors='coerce')

# Count new missing values
print("New missing values after conversion:")
print(df[['age', 'weight']].isnull().sum())

# Convert insurance_provider to category
df['insurance_provider'] = df['insurance_provider'].astype('category')

print("\nData Types After Conversion:")
print(df.dtypes)
```

## === TASK 3: Handle Missing Values ===

```

# Fill numeric columns with median
df['age'].fillna(df['age'].median(), inplace=True)
df['weight'].fillna(df['weight'].median(), inplace=True)
df['blood_pressure'].fillna(df['blood_pressure'].median(), inplace=True)

# Fill medication with mode
df['medication'].fillna(df['medication'].mode()[0], inplace=True)

# Fill insurance_provider with 'Unknown'
df['insurance_provider'].fillna('Unknown', inplace=True)

print("Remaining Missing Values:")
print(df.isnull().sum())
```

## === TASK 4: Handle Duplicates ===

```

print("Shape Before Removing Duplicates:", df.shape)

print("\nDuplicate Rows (All Columns):")
print(df[df.duplicated(keep=False)])

print("\nDuplicates based on patient_id:")
print(df[df.duplicated(subset=['patient_id'], keep=False)])

df = df.drop_duplicates(subset=['patient_id'], keep='first')

print("\nShape After Removing Duplicates:", df.shape)
```

## === TASK 5: Complete Workflow with Verification ===

```

# Reload original DataFrame
df_original = pd.DataFrame(data)
df_clean = df_original.copy()

print("Shape Before Cleaning:", df_clean.shape)
print("Missing Before Cleaning:\n", df_clean.isnull().sum())
print("Duplicates Before Cleaning:", df_clean.duplicated(subset=['patient_id']).sum())
print("Data Types Before Cleaning:\n", df_clean.dtypes)

# Cleaning Process
df_clean['age'] = pd.to_numeric(df_clean['age'], errors='coerce')
df_clean['weight'] = pd.to_numeric(df_clean['weight'], errors='coerce')
df_clean['insurance_provider'] = df_clean['insurance_provider'].astype('category')

df_clean['age'].fillna(df_clean['age'].median(), inplace=True)
df_clean['weight'].fillna(df_clean['weight'].median(), inplace=True)
df_clean['blood_pressure'].fillna(df_clean['blood_pressure'].median(), inplace=True)
df_clean['medication'].fillna(df_clean['medication'].mode()[0], inplace=True)
df_clean['insurance_provider'].fillna('Unknown', inplace=True)

df_clean = df_clean.drop_duplicates(subset=['patient_id'], keep='first')

print("\nShape After Cleaning:", df_clean.shape)
print("Missing After Cleaning:\n", df_clean.isnull().sum())
print("Duplicates After Cleaning:", df_clean.duplicated(subset=['patient_id']).sum())
print("Data Types After Cleaning:\n", df_clean.dtypes)
```
