## Import Required Libraries

```

import pandas as pd
import numpy as np
```

## === TASK 1: Data Preparation and Missing Value Handling ===

### Create Students, Enrollments, and Scores DataFrames

```

students_data = {
    'student_id': [101, 102, 103, 104, 105, 106, 107],
    'name': ['Alice', 'Bob', None, 'David', 'Emma', 'Frank', 'Grace'],
    'email': ['alice@email.com', 'bob@email.com', 'charlie@email.com', None, 'emma@email.com', 'frank@email.com', 'grace@email.com'],
    'city': ['Mumbai', 'Delhi', 'Bangalore', 'Mumbai', None, 'Chennai', 'Delhi']
}

enrollments_data = {
    'student_id': [101, 102, 103, 105, 108, 109],
    'course_name': ['Python', 'Data Science', 'Python', 'Machine Learning', 'AI', 'Python'],
    'enrollment_date': ['2024-01-15', '2024-01-20', '2024-02-01', '2024-02-10', '2024-02-15', '2024-03-01']
}

scores_data = {
    'student_id': [101, 102, 104, 105, 106],
    'exam_score': [85, 92, 78, 88, 95]
}

students = pd.DataFrame(students_data)
enrollments = pd.DataFrame(enrollments_data)
scores = pd.DataFrame(scores_data)

students
```

### Null Value Count and Percentage

```

null_counts = students.isnull().sum()
null_percent = (null_counts / len(students)) * 100

for col in students.columns:
    print(f"Column: {col}, Nulls: {null_counts[col]} ({null_percent[col]:.2f}%)")
```

### Fill Missing City and Drop Missing Names

```

students['city'].fillna('Unknown', inplace=True)
students_cleaned = students.dropna(subset=['name'])

students_cleaned
```

## === TASK 2: Join Operations ===

### Inner Join

```

inner_join = pd.merge(students_cleaned, enrollments, on='student_id', how='inner')
inner_join
```

### Left Join

```

left_join = pd.merge(students_cleaned, enrollments, on='student_id', how='left')
left_join
```

### Right Join

```

right_join = pd.merge(students_cleaned, enrollments, on='student_id', how='right')
right_join
```

### Full Outer Join with Indicator

```

outer_join = pd.merge(students_cleaned, enrollments, on='student_id', how='outer', indicator=True)
outer_join

print("\nMerge Source Distribution:")
print(outer_join['_merge'].value_counts())
```

## === TASK 3: Lookup and Automation ===

### Lookup Operation using map()

```

score_dict = dict(zip(scores['student_id'], scores['exam_score']))

students_cleaned['exam_score'] = students_cleaned['student_id'].map(score_dict)

students_cleaned[['student_id', 'name', 'exam_score']]
```

### Adding Scores using merge()

```

merge_scores = pd.merge(students_cleaned, scores, on='student_id', how='left')
merge_scores[['student_id', 'name', 'exam_score']]
```

### Automation Function

```

def auto_merge(df1, df2, join_type, key_column):
    merged_df = pd.merge(df1, df2, on=key_column, how=join_type)
    return {
        'result_df': merged_df,
        'row_count': len(merged_df),
        'join_type': join_type
    }

# Test function
test_inner = auto_merge(students_cleaned, enrollments, 'inner', 'student_id')
test_left = auto_merge(students_cleaned, enrollments, 'left', 'student_id')

print("Inner Join Rows:", test_inner['row_count'])
print("Left Join Rows:", test_left['row_count'])

test_inner['result_df'].head()
```
